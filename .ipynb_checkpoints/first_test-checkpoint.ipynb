```python
import pandas as pd
import numpy as np
```

```python
epoch = 10
hidden_layer_size = 200
hidden_layers = 3

np.random.seed(10)
```

```python
def import_data():
    
    train = pd.read_csv("train.csv")
    X_test = pd.read_csv("test.csv")
    
    y_train = (train["class"])
    X_train = train.drop(columns = "class")
    
    return X_train, X_test, y_train
    
```

```python
X_train_, X_test_, y_train = import_data()
```

```python
features_to_keep = ["freq", "premier_pay", "an", 
                    "depot", "marque",
                    "prix_auto", "pour_depot", "km", 
                    "sexe","ristourne", "cheque_dealer",
                    "sommes_des_couts", "terme"]
                    
X_train = X_train_[features_to_keep]
X_test = X_test_[features_to_keep]
```

```python
dum_marque_train = (pd.get_dummies(X_train["marque"])).fillna(0)
dum_marque_test = (pd.get_dummies(X_test["marque"])).fillna(0)
```

```python
X_train = (pd.concat([X_train.drop(columns = ["marque"]), dum_marque_train], axis = 1))
X_test = (pd.concat([X_test.drop(columns = ["marque"]), dum_marque_test], axis = 1))
```

```python
good_col = []
for col in X_train.columns.values:
    if col in X_test.columns.values:
        good_col.append(col)

X_train = X_train[good_col]
```

```python
new_freq_train = X_train["freq"].replace(['Par jour(s)', 'Mensuelle'], [0,1])
new_freq_test = X_test["freq"].replace(['Par jour(s)', 'Mensuelle'], [0,1])

X_train = (pd.concat([X_train.drop(columns = ["freq"]), new_freq_train], axis = 1))
X_test = (pd.concat([X_test.drop(columns = ["freq"]), new_freq_test], axis = 1))
```

```python
print("X_train", (X_train.shape))
print("X_test", (X_test.shape))
```

```python
def nonlin(x, deriv=False):
    if(deriv==True):
        return x*(1-x)

    return 1/(1+np.exp(-x))
```

```python
synA = 2*np.random.random((37,hidden_layer_size)) - 1
synB = 2*np.random.random((hidden_layer_size,1)) - 1

def train_model():
    global synA
    global synB
    
    a = X_train
    b = y_train

    list_of_error_l1=[]
    list_of_error_l2=[]

    for j in range(600):

        l0 = a
        l1 = nonlin(np.dot(l0, synA))
        l2 = nonlin(np.dot(l1, synB))

        l2_error = b - l2
        if (j % 10 == 0):
            print ("Error: " + str(np.mean(np.abs(l2_error))))

        l2_delta = l2_error*nonlin(l2, deriv=True)
        l1_error = l2_delta.dot(synB.T)
        l1_delta = l1_error * nonlin(l1,deriv=True)

        #update weights
        synB += l1.T.dot(l2_delta)
        synA += l0.T.dot(l1_delta)

        list_of_error_l1.append(str(np.mean(np.abs(l1_error))))
        list_of_error_l2.append(str(np.mean(np.abs(l2_error))))
```

```python
train_model()
```
